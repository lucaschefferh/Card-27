```python
#Importação das bibliotecas necessárias
import tensorflow as tf
import tensorflow.keras.backend as K
import os
import random
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.layers import Lambda
from tensorflow.keras.callbacks import LearningRateScheduler, Callback

print(tf.config.list_physical_devices('GPU'))
```

```python
#Carregamento da base de dados CIFAR-10
from tensorflow.keras.datasets import cifar10

(x_train, y_train), (x_test, y_test) = cifar10.load_data()
#Reduz o conjunto de treinamento pela metade 
x_train = x_train[:25000]
y_train = y_train[:25000]

assert x_train.shape == (25000, 32, 32, 3)
assert x_test.shape == (10000, 32, 32, 3)

#Classes do CIFAR-10
classes = ['airplane', 'automobile', 'bird', 'cat', 'deer',
           'dog', 'frog', 'horse', 'ship', 'truck']

#Visualização de algumas imagens aleatórias
plt.figure(figsize=(12, 12))
rnd_samples = np.random.choice(25000, 9)

for i in range(9):
    plt.subplot(3, 3, i+1)
    plt.imshow(x_train[rnd_samples[i]])
    plt.title(f"Classe: {classes[y_train[rnd_samples[i]][0]]}")
    plt.axis('off')

plt.show()
```

```python
#Pré-processamento dos dados
dataset = np.concatenate([x_train, x_test], axis=0)
dataset = dataset.astype("float32") / 255.0  # Normalização para [0,1]

#Redimensionar as imagens para 64x64 para ter mais detalhes
dataset = tf.image.resize(dataset, (64, 64)).numpy()
```

```python
from tensorflow import keras
from tensorflow.keras import layers

#Criação da camada de amostragem
class SamplingLayer(layers.Layer):    
    def call(self, inputs):
        z_mean, z_log_var = inputs
        batch = tf.shape(z_mean)[0]
        dim = tf.shape(z_mean)[1]
        epsilon = tf.random.normal(shape=(batch, dim))
        return z_mean + tf.exp(0.5 * z_log_var) * epsilon
```

```python
#Definição do bloco residual
def residual_block(x, filters, kernel_size=3):
    y = layers.Conv2D(filters, kernel_size, padding='same')(x)
    y = layers.BatchNormalization()(y)
    y = layers.LeakyReLU()(y)
    y = layers.Conv2D(filters, kernel_size, padding='same')(y)
    y = layers.BatchNormalization()(y)
    out = layers.Add()([x, y])
    return layers.LeakyReLU()(out)

#Construção do Encoder
def build_encoder(latent_dim, encoder_inputs):
    #Arquitetura 
    x = layers.Conv2D(32, 3, strides=2, padding="same")(encoder_inputs)
    x = layers.BatchNormalization()(x)
    x = layers.LeakyReLU()(x)
    
    #Primeiro bloco residual
    x = residual_block(x, 32)
    x = layers.Conv2D(64, 3, strides=2, padding="same")(x)
    x = layers.BatchNormalization()(x)
    x = layers.LeakyReLU()(x)
    
    #Segundo bloco residual
    x = residual_block(x, 64)
    x = layers.Conv2D(128, 3, strides=2, padding="same")(x)
    x = layers.BatchNormalization()(x)
    x = layers.LeakyReLU()(x)
    
    #Terceiro bloco residual
    x = residual_block(x, 128)
    
    #Flatten e Dense layers com dropout
    x = layers.Flatten()(x)
    x = layers.Dense(512)(x)
    x = layers.BatchNormalization()(x)
    x = layers.LeakyReLU()(x)
    x = layers.Dropout(0.3)(x)
    
    #Camadas latentes
    z_mean = layers.Dense(latent_dim, name="z_mean")(x)
    z_log_var = layers.Dense(latent_dim, name="z_log_var")(x)
    z = SamplingLayer()([z_mean, z_log_var])
    
    return keras.Model(encoder_inputs, [z_mean, z_log_var, z], name="encoder")

#Criação do encoder com dimensão latente maior
latent_dim = 128
encoder_inputs = keras.Input(shape=(64, 64, 3))
encoder = build_encoder(latent_dim, encoder_inputs)
encoder.summary()
```

```python
#Construção do Decoder
def build_decoder(latent_inputs):
    #Dimensão maior para ter mais detalhes
    x = layers.Dense(8 * 8 * 256)(latent_inputs)
    x = layers.BatchNormalization()(x)
    x = layers.LeakyReLU()(x)
    x = layers.Reshape((8, 8, 256))(x)
    
    #Primeira camada de upsampling com bloco residual
    x = layers.Conv2DTranspose(128, 3, strides=2, padding="same")(x)
    x = layers.BatchNormalization()(x)
    x = layers.LeakyReLU()(x)
    x = residual_block(x, 128)
    
    #Segunda camada de upsampling com bloco residual
    x = layers.Conv2DTranspose(64, 3, strides=2, padding="same")(x)
    x = layers.BatchNormalization()(x)
    x = layers.LeakyReLU()(x)
    x = residual_block(x, 64)
    
    #Terceira camada de upsampling com bloco residual
    x = layers.Conv2DTranspose(32, 3, strides=2, padding="same")(x)
    x = layers.BatchNormalization()(x)
    x = layers.LeakyReLU()(x)
    x = residual_block(x, 32)
    
    #Camada final para gerar a imagem
    decoder_outputs = layers.Conv2D(3, 3, activation="sigmoid", padding="same")(x)
    
    return keras.Model(latent_inputs, decoder_outputs, name="decoder")

#Criação do decoder
latent_inputs = keras.Input(shape=(latent_dim,))
decoder = build_decoder(latent_inputs)
decoder.summary()
```

```python
#Callback para warm-up do KL
class KLAnnealingCallback(Callback):
    def __init__(self, warmup_epochs=10, max_kl_weight=1.0):
        super().__init__()
        self.warmup_epochs = warmup_epochs
        self.max_kl_weight = max_kl_weight
        
    def on_epoch_begin(self, epoch, logs=None):
        K.set_value(self.model.kl_weight, 
                   min(self.max_kl_weight, epoch / self.warmup_epochs * self.max_kl_weight))

#Função de perda de reconstrução usando MSE e perceptual loss
def reconstruction_loss(data, reconstructed):
    #MSE loss
    mse_loss = tf.reduce_mean(
        tf.reduce_sum(
            tf.keras.losses.mse(data, reconstructed),
            axis=(1, 2)
        )
    )
    
    #Perceptual loss simplificado (diferença entre features de baixo nível)
    perceptual_weight = 0.1
    data_features = tf.image.sobel_edges(data)
    reconstructed_features = tf.image.sobel_edges(reconstructed)
    perceptual_loss = tf.reduce_mean(
        tf.reduce_sum(
            tf.keras.losses.mse(data_features, reconstructed_features),
            axis=(1, 2, 3)
        )
    )
    
    return mse_loss + perceptual_weight * perceptual_loss

#Função de perda KL Divergence com peso dinâmico
def kl_divergence_loss(z_mean, z_log_var):
    kl_loss = -0.5 * tf.reduce_mean(
        tf.reduce_sum(
            1 + z_log_var - tf.square(z_mean) - tf.exp(z_log_var),
            axis=1
        )
    )
    return kl_loss

#Cálculo da perda total
def calc_total_loss(data, reconstructed, z_mean, z_log_var, kl_weight):
    reconstruction = reconstruction_loss(data, reconstructed)
    kl = kl_divergence_loss(z_mean, z_log_var)
    return reconstruction, kl, reconstruction + kl_weight * kl
```

```python
#Implementação da classe VAE
class VAE(keras.Model):
    def __init__(self, encoder, decoder, **kwargs):
        super(VAE, self).__init__(**kwargs)
        self.encoder = encoder
        self.decoder = decoder
        self.kl_weight = K.variable(0.0)  #Começa com 0 para warm-up
        
        #Métricas para acompanhamento
        self.total_loss_tracker = keras.metrics.Mean(name="total_loss")
        self.reconstruction_loss_tracker = keras.metrics.Mean(name="reconstruction_loss")
        self.kl_loss_tracker = keras.metrics.Mean(name="kl_loss")
        self.val_total_loss_tracker = keras.metrics.Mean(name="val_total_loss")
        
    @property
    def metrics(self):
        return [
            self.total_loss_tracker,
            self.reconstruction_loss_tracker,
            self.kl_loss_tracker,
            self.val_total_loss_tracker
        ]
    
    def train_step(self, data):
        with tf.GradientTape() as tape:
            #Forward pass
            z_mean, z_log_var, z = self.encoder(data)
            reconstruction = self.decoder(z)
            
            #Cálculo das perdas
            reconstruction_loss, kl_loss, total_loss = calc_total_loss(
                data, reconstruction, z_mean, z_log_var, self.kl_weight
            )
        
        #Backward pass
        grads = tape.gradient(total_loss, self.trainable_weights)
        
        #Gradient clipping para estabilidade
        clipped_grads, _ = tf.clip_by_global_norm(grads, clip_norm=5.0)
        self.optimizer.apply_gradients(zip(clipped_grads, self.trainable_weights))
        
        #Atualização das métricas
        self.total_loss_tracker.update_state(total_loss)
        self.reconstruction_loss_tracker.update_state(reconstruction_loss)
        self.kl_loss_tracker.update_state(kl_loss)
        
        return {
            "loss": self.total_loss_tracker.result(),
            "reconstruction_loss": self.reconstruction_loss_tracker.result(),
            "kl_loss": self.kl_loss_tracker.result(),
        }
    
    def test_step(self, data):
        #Forward pass
        z_mean, z_log_var, z = self.encoder(data)
        reconstruction = self.decoder(z)
        
        #Cálculo das perdas
        _, _, total_loss = calc_total_loss(
            data, reconstruction, z_mean, z_log_var, self.kl_weight
        )
        
        #Atualização da métrica de validação
        self.val_total_loss_tracker.update_state(total_loss)
        
        return {"val_loss": self.val_total_loss_tracker.result()}
```

```python
#Learning rate scheduler
def cosine_decay_schedule(epoch, lr):
    initial_lr = 0.001
    decay_epochs = 40
    min_lr = 1e-6
    
    if epoch < 5:  #Warm-up period
        return initial_lr * ((epoch + 1) / 5)
    else:
        #Cosine decay
        progress = (epoch - 5) / (decay_epochs - 5)
        cosine_decay = 0.5 * (1 + np.cos(np.pi * progress))
        return min_lr + (initial_lr - min_lr) * cosine_decay

#Preparação dos callbacks
callbacks = [
    KLAnnealingCallback(warmup_epochs=10, max_kl_weight=1.0),
    LearningRateScheduler(cosine_decay_schedule),
    tf.keras.callbacks.EarlyStopping(
        monitor='val_loss',
        patience=5,
        restore_best_weights=True
    )
]

#Separação dos dados em treino e validação
val_split = 0.1
val_size = int(len(dataset) * val_split)
train_data = dataset[:-val_size]
val_data = dataset[-val_size:]

#Treinamento do modelo
vae = VAE(encoder, decoder)
vae.compile(
    optimizer=keras.optimizers.AdamW(
        learning_rate=0.001,
        weight_decay=0.001
    )
)

#Treinamento com validação
history = vae.fit(
    train_data,
    epochs=10,
    batch_size=16,
    validation_data=(val_data, None),
    callbacks=callbacks
)
```

```python
#Visualização das métricas durante o treinamento
plt.figure(figsize=(15, 5))

#Plot das perdas
plt.subplot(1, 3, 1)
plt.plot(history.history['loss'], label='Perda Total (Treino)')
plt.plot(history.history['reconstruction_loss'], label='Perda Reconstrução')
plt.plot(history.history['kl_loss'], label='Perda KL')
plt.title('Evolução das Perdas')
plt.xlabel('Época')
plt.ylabel('Perda')
plt.legend()

#Plot da perda KL isolada
plt.subplot(1, 3, 2)
plt.plot(history.history['kl_loss'], label='Perda KL')
plt.title('Perda KL Isolada')
plt.xlabel('Época')
plt.ylabel('Perda KL')
plt.legend()

#Plot do learning rate
plt.subplot(1, 3, 3)
lrs = [cosine_decay_schedule(epoch, None) for epoch in range(len(history.history['loss']))]
plt.plot(lrs, label='Learning Rate')
plt.title('Learning Rate Schedule')
plt.xlabel('Época')
plt.ylabel('Learning Rate')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Geração de novas imagens com interpolação no espaço latente
def interpolate_points(p1, p2, n_steps=10):
    ratios = np.linspace(0, 1, num=n_steps)
    vectors = []
    for ratio in ratios:
        v = (1.0 - ratio) * p1 + ratio * p2
        vectors.append(v)
    return np.array(vectors)

#Gerar pontos aleatórios no espaço latente
n_samples = 4
n_steps = 10

z_samples = np.random.normal(0, 1, size=(n_samples, latent_dim))
num_sequences = int(n_samples * (n_samples - 1) / 2)

#Criar grade de interpolações
plt.figure(figsize=(n_steps * 1.5, num_sequences * 1.5))

sequence_counter = 0

for i in range(n_samples):
    for j in range(i + 1, n_samples):
        #Interpolação entre dois pontos
        interpolated = interpolate_points(z_samples[i], z_samples[j], n_steps=n_steps)
        decoded = vae.decoder.predict(interpolated)
        
        sequence_counter += 1
        
        #Plotar interpolação
        for k in range(len(decoded)):
            plot_index = (sequence_counter - 1) * n_steps + (k + 1)
            
            plt.subplot(num_sequences, n_steps, plot_index)
            plt.imshow(decoded[k])
            plt.axis('off')

plt.suptitle('Interpolações no Espaço Latente', size=16)
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()

#Geração de imagens aleatórias
z = np.random.normal(0, 1, size=(25, latent_dim))
generated_images = vae.decoder.predict(z)

plt.figure(figsize=(15, 15))
for i in range(25):
    plt.subplot(5, 5, i+1)
    plt.imshow(generated_images[i])
    plt.axis('off')
plt.suptitle('Imagens Geradas pelo VAE', size=16)
plt.show()
```
